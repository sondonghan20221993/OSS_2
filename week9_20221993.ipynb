```python
class HauntedBus:
    """유령 승객이 출몰하는 버스 모델"""

    def __init__(self, passengers=[]):
        # p[] 가 intepret 단계에서 생성되며 모든 인스턴스들이 같은 대상을 참조하게되어 오류 발생
        self.passengers = passengers

    def pick(self, name):
        self.passengers.append(name)

    def drop(self, name):
        self.passengers.remove(name)
```

```python
bus1 = HauntedBus(['Alice', 'Bill'])
bus1.passengers
```

```python
bus1.pick('Charlie')
bus1.drop('Alice')
bus1.passengers
```

```python
bus2 = HauntedBus()
bus2.pick('Carrie')
bus2.passengers
```

```python
bus3 = HauntedBus()
bus3.passengers # 초기값이 더이상 empty가 아님!
```

```python
bus3.pick('Dave')
bus2.passengers
```

```python
bus2.passengers is bus3.passengers
```

```python
class TwilightBus:
    """승객이 사라지게 만드는 버스 모델"""

    def __init__(self, passengers=None):
        if passengers is None:
            self.passengers = []
        else:
            self.passengers = passengers # 사본 생성이 아니라 인수를 별명으로 사용

    def pick(self, name):
        self.passengers.append(name)

    def drop(self, name):
        self.passengers.remove(name)
```

```python
basketball_team = ['Sue', 'Tina', 'Maya', "Diana", "Pat"]
bus = TwilightBus(basketball_team)
bus.drop('Tina')
bus.drop('Pat')
basketball_team # 선수단이 변경되어 버림
```
